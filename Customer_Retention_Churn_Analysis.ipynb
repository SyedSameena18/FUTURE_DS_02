DATA VISUALIZATION

1. CUSTOMER CHURN COUNT - The company has a significant number of customers leaving the service, indicating customer retention challenges.





```python
sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Count")
plt.show()
```

2. CONTRACT vs CHURN ANALYSIS -
Customers with month-to-month contracts show the highest churn rate, while long-term contracts help retain customers better.

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.xticks(rotation=45)
plt.show()
```

3. PAYMENT METHOD vs CHURN -
Customers using electronic check payment methods tend to churn more compared to automatic payment methods.

```python
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.xticks(rotation=90)
plt.title("Payment Method vs Churn")
plt.show()
```

4.INTERNET SERVICE vs CHURN -
Fiber optic internet users show a relatively higher churn rate, indicating possible dissatisfaction or pricing concerns.

```python
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Internet Service vs Churn")
plt.show()
```

5. TENURE DISTRIBUTION -
Customers with shorter tenure are more likely to churn, while long-term customers tend to remain loyal.

```python
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack')
plt.title("Customer Tenure Distribution")
plt.show()
```

KPI METRICS -

1. The company experiences a notable customer churn rate, indicating retention challenges.

2. Customers with month-to-month contracts are more likely to leave compared to long-term contract users.

3. Electronic check payment users demonstrate higher churn behavior.

4. Customers with shorter tenure show lower loyalty and higher risk of churn.

5. Fiber optic internet customers display relatively higher churn patterns, suggesting possible service or pricing dissatisfaction.





1. TOTAL CUSTOMERS KPI

```python
total_customers = df.shape[0]
print("Total Customers:", total_customers)
```

2. CHURN COUNT KPI

```python
churn_count = df[df['Churn'] == 'Yes'].shape[0]
print("Churned Customers:", churn_count)
```

3. CHURN PERCENTAGE KPI

```python
churn_rate = (churn_count / total_customers) * 100
print("Churn Rate: {:.2f}%".format(churn_rate))
```

4. AVERAGE MONTHLY CHARGES

```python
avg_monthly = df['MonthlyCharges'].mean()
print("Average Monthly Charges:", round(avg_monthly, 2))
```

5. AVERAGE TENURE

```python
avg_tenure = df['tenure'].mean()
print("Average Customer Tenure:", round(avg_tenure, 2))
```

BUSINESS INSIGHTS -

1. The company is experiencing a significant customer churn rate, indicating retention challenges.

2. Customers with month-to-month contracts have the highest likelihood of churning compared to customers with long-term contracts.

3. Customers using electronic check payment methods show higher churn behavior than customers using automatic payment methods.

4. Fiber optic internet users demonstrate relatively higher churn patterns, suggesting possible concerns related to pricing, service quality, or customer satisfaction.

5. Customers with shorter tenure are more likely to leave the service, while long-term customers tend to remain loyal.

6. Long-term contracts and automatic payment methods appear to positively influence customer retention.

7. Customer retention strategies should primarily focus on newly acquired customers and high-risk service categories.

BUSINESS RECOMMENDATIONS -

1. Introduce discounts and loyalty rewards for long-term subscriptions.

2. Encourage customers to switch from month-to-month plans to yearly contracts.

3. Improve customer support and pricing strategies for fiber optic users.

4. Promote automatic payment methods to improve retention.

5. Create targeted retention campaigns for new customers during early tenure periods.
